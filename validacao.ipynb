```python
### MULTITHREADER ###

def runner(list, function):
    threads= []
    with ThreadPoolExecutor(max_workers=20) as executor:
        for item in list:
          
            threads.append(executor.submit(function, item))
           
        for task in as_completed(threads):

            if task.result() == None:

                continue
            
            else:

                print(task.result())
```

```python
## EXTRAIR OS CÓDIGOS DE TODOS OS ANÚNCIOS DA CONTA ##


def todos_anuncios (accountname, header):
  """
  Essa função extrai todos os anúncios relacionados a conta do MeLi,
  criando a lista 'codigos_anuncios'
  """

  print ("Rodando...")

  global codigos_anuncios

  codigos_anuncios = []

  try:

    url = f"https://portal.vtexcommercestable.com.br/api/meliintegration/stockkeepingunit/scan?an={accountname}&lt=gold_pro&limit=100"

    response = requests.request("GET", url, headers=header)

    codigos_anuncios.append(response.json()['results'])

  except:

    url = f"https://portal.vtexcommercestable.com.br/api/meliintegration/stockkeepingunit/scan?an={accountname}&lt=gold_special&limit=100"

    response = requests.request("GET", url, headers=header)

    codigos_anuncios.append(response.json()['results'])

  while len(response.json()['results']) > 0:

      scrollid = response.json()['scroll_id']

      try:

        url = f"https://portal.vtexcommercestable.com.br/api/meliintegration/stockkeepingunit/scan?an={accountname}&lt=gold_pro&limit=100&scrollid={scrollid}"

        response = requests.request("GET", url, headers=header)

        codigos_anuncios.append(response.json()['results'])

      except:

        url = f"https://portal.vtexcommercestable.com.br/api/meliintegration/stockkeepingunit/scan?an={accountname}&lt=gold_special&limit=100&scrollid={scrollid}"

        response = requests.request("GET", url, headers=header)

        codigos_anuncios.append(response.json()['results'])
    
      print(len(codigos_anuncios))

  codigos_anuncios = list(chain(*codigos_anuncios))

  print(f"Todos os códigos dos {len(codigos_anuncios)} anúncios da conta {accountname} no MeLi foram extraídos com sucesso! A lista está armazenada em 'codigos_anuncios'.")

  return codigos_anuncios
```

```python
## EXTRAIR INFOS DE TODOS OS ANÚNCIOS ##

status = []
id_anuncio = []
tipo_anuncio = []
id_variacao = []
seller_custom_field = []
catalog_listing = []

b=0
start_time = time.time()
total_anuncios = len(codigos_anuncios)

def validacao_migracao(anuncio):

    url = f"https://portal.vtexcommercestable.com.br/api/meliintegration/item/{anuncio}?an={accountname}"

    listing_info = requests.request("GET", url, headers=header)

    global b
    b = b+1
    now = time.time()
    percent_complete = b/total_anuncios

    print(f'{anuncio} : {b} anúncios extraídos - {"{:.0%}".format(percent_complete)} completo (Faltam {str( round(((now-start_time)/percent_complete)-(now-start_time)))} segundos)')

    if listing_info.json()['status'] == 'under_review':

        status.append(listing_info.json()['status'])
        id_anuncio.append(listing_info.json()['id'])
        tipo_anuncio.append(listing_info.json()['listing_type_id'])
        id_variacao.append("0")
        seller_custom_field.append("0")
        catalog_listing.append("0")


    elif listing_info.json()['variations'] == []:

        status.append(listing_info.json()['status'])
        id_anuncio.append(listing_info.json()['id'])
        tipo_anuncio.append(listing_info.json()['listing_type_id'])
        id_variacao.append("sem variação")
        catalog_listing.append(listing_info.json()['catalog_listing'])

        try:
            seller_custom_field.append(listing_info.json()['seller_custom_field'])
        except:
            seller_custom_field.append('SEM')

        
    else:

        for k in range(len(listing_info.json()['variations'])):

            status.append(listing_info.json()['status'])
            id_anuncio.append(listing_info.json()['id'])
            tipo_anuncio.append(listing_info.json()['listing_type_id'])
            id_variacao.append(listing_info.json()['variations'][k]['id'])
            catalog_listing.append(listing_info.json()['catalog_listing'])

            try:
                seller_custom_field.append(listing_info.json()['variations'][k]['seller_custom_field'])
            except:
                seller_custom_field.append('SEM')

runner(codigos_anuncios, validacao_migracao)
    
infos_meli = pd.DataFrame({'status':status,'tipo_anuncio':tipo_anuncio,'codigo_anuncio':id_anuncio, 'id_variacao':id_variacao, 'seller_custom_field':seller_custom_field, 'catalog_listing':catalog_listing})

infos_meli.to_csv(f'validation_listings_{accountname}.csv', encoding='ISO-8859-1')


print(f'Acabou! O processo finalizou com sucesso e a planilha "validation_listings_{accountname}.csv" está na pasta de destino')
```
